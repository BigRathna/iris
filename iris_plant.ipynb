```python
# pattern recodnition for iris plants

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
# loading the iris dataset
iris = pd.read_csv(r'iris\iris.data',delimiter=',',header=None)

# naming the columns
iris.columns = ['sepal_length','sepal_width','petal_length','petal_width','class']

# checking the data
print(iris.head())
```

```python
# preparing the model
X = iris.drop(['class'],axis=1)
y = iris['class']

# splitting the data
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)

```

```python
#training the model

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
```

```python
# predicting the model
y_pred = knn.predict(X_test)

# checking the accuracy
print(accuracy_score(y_test,y_pred))
```

```python
# confusion matrix
print(confusion_matrix(y_test,y_pred))


# classification report
print(classification_report(y_test,y_pred))

```

```python
#visualizing the results
sns.pairplot(iris,hue='class',palette='Dark2')
plt.show()
```

```python
#predicting the class of a new data with input form user
sepal_length = float(input('Enter sepal length: '))
sepal_width = float(input('Enter sepal width: '))
petal_length = float(input('Enter petal length: '))
petal_width = float(input('Enter petal width: '))
new_data = np.array([[sepal_length,sepal_width,petal_length,petal_width]])
new_pred = knn.predict(new_data)
print('Predicted class: ',new_pred[0])
```
